```python
def add_company(inpt):
    if inpt == 'Samsung Note5 64GB' or inpt == 'Samsung Note5 128GB' or inpt == 'Samsung S5 Neo':
        return 'Samsung Note5'
    else:
        return inpt
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25)
```

```python
X_train.shape,X_test.shape
```

```python
def model_tra(model):
    model.fit(X_train, y_train)
    acc = model.score(X_test, y_test)
    print(str(model)+ ' --> ' +str(acc))
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
model_tra(lr)

from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor()
model_tra(rfr)

```

```python
from sklearn.model_selection import GridSearchCV

parameters = {'n_estimators':[10, 50, 100],
              'criterion':['squared_error','absolute_error','poisson']}

grid_object = GridSearchCV(estimator=rfr, param_grid=parameters)

grid_fit = grid_object.fit(X_train, y_train)

best_modelpara = grid_fit.best_estimator_
best_modelpara

```

```python
best_modelpara.score(X_test, y_test)
```

```python
import pickle
with open('money_predictor.pickle','wb') as file:
    pickle.dump(best_modelpara,file)
```
